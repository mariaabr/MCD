### LAB: DNN with keras and tensorflow

**Example 1: MNIST dataset (Digit recognition dataset)**

MNIST is available as a keras dataset (mnist)

Inputs: images of 28 x 28 pixels

Output: class representing the digit (10 classes, digitos 0-9)

60k images for training and 10k images for test


```python
from tensorflow.keras.datasets import mnist
import numpy as np
import matplotlib.pyplot as plt

(train_images, train_labels), (test_images, test_labels) = mnist.load_data()

# Check data dimensions
print("Train Images Shape:", train_images.shape)
print("Train Labels Shape:", train_labels.shape)
print("Test Images Shape:", test_images.shape)
print("Test Lables Shape:", test_labels.shape)
```

### Data pre-processing

**Classification problem => Convert outputs into categorical variables**

```python
from tensorflow.keras.utils import to_categorical

# Reshape to make the images into 1D vector  for each example
train_images = train_images.reshape(train_images.shape[0], -1)
print("Train Images shape:", train_images.shape)

# Standardize values /255
train_images = train_images.astype('float32') / 255

# Convert outputs into categorical variables 
train_labels = to_categorical(train_labels)
print("Train Labels shape:", train_labels.shape)

# Do the same for the test data
test_images = test_images.reshape(test_images.shape[0], -1)
print("Test Images shape:", test_images.shape)

# Standardize values /255
test_images = test_images.astype('float32') / 255

# Convert outputs into categorical variables 
test_labels = to_categorical(test_labels)
print("Test Labels shape:", test_labels.shape)
```

### Model structure (feedforward DNN)

```python
from tensorflow.keras import models
from tensorflow.keras import layers

network = models.Sequential()

#One hidden layer with 512 neurons (RelU)
network.add(layers.Dense(512, activation='relu', 
                         input_shape=(28 * 28,)))

# Output layer with 10 neurons
# (softmax = 1 neuron for each class; one-hot encoding )
network.add(layers.Dense(10, activation='softmax'))
```

### DNN training

```python
network.compile(optimizer='rmsprop',
                loss='categorical_crossentropy',
                metrics=['accuracy'])

# train the model (fit) with batch_size=128 
history = network.fit(train_images, train_labels, epochs=10, batch_size=128)
```

```python
test_preds = network.predict(test_images)

# test_preds contains the class probabilities, what is its dimension ?
print("Test predicitions shape:", test_preds.shape)
```

```python
# (choose the class with max probabilities)
test_classes = np.argmax(network.predict(test_images), axis=-1)

# what is the dimension of test_classes ?
print("Test classes shape:", test_preds.shape)
```

**3) *network.evaluate* - evaluate the trained model for test and train images. Compare the final loss and accuracy values**

```python
train_loss, train_acc = network.evaluate(train_images, train_labels) # less loss more accuracy
test_loss, test_acc = network.evaluate(test_images, test_labels)
```

### Example 2:  DNN classifier for IMDB dataset

Dataset with texts of reviews in IMDB about movies; classified into 2 
classes: positive or negative

25k reviews for training + 25k for test; 

balanced – 50% positive and negative examples

IMDB is available as a keras dataset (imdb)

We will only consider the most common 10k words

```python
from tensorflow.keras.datasets import imdb

# Use text dictionary with num_words=10000
(train_data, train_labels), (test_data, test_labels) = imdb.load_data(num_words=10000)
```

```python
# Check dimensions of train and test data, labels
print("Train Data Shape:", train_data.shape)
print("Train Labels Shape:", train_labels.shape)
print("Test Data Shape:", test_data.shape)
print("Test Lables Shape:", test_labels.shape)

# what is the content of train_data and train_labels ? 
print("The train_data has the observations to predict the labels and train_labels are the labels of the observations, the output.")

# Print the content of the first review 
print("Content of the first review:", train_data[0])

# The number of words in each review are different 
# How many words are in review 1, review 4
print("Number of words of review 1:", len(train_data[1]))
print("Number of words of review 4:", len(train_data[3]))

# ANSWER: 218 words / #550 words 

# What is the class of the first review
print("Class of the first review:", train_labels[0])
```

**Transform the word indices back into words** 

```python
print(max([max(sequence) for sequence in train_data]))

word_index = imdb.get_word_index()
reverse_word_index = dict([(value, key) for (key, value) in word_index.items()])
decoded_review = ' '.join([reverse_word_index.get(i - 3, '?') for i in train_data[2]])
print(decoded_review)
```

```python
def vectorize_sequences(sequences, dimension=10000):
    results = np.zeros((len(sequences), dimension))
    for i, sequence in enumerate(sequences):
        results[i, sequence] = 1.
    return results

# One-hot encoding of the reviews (binary vectors)

x_train = vectorize_sequences(train_data)
x_test = vectorize_sequences(test_data)

# Output to binary (0 / 1 Classes) 
y_train = np.asarray(train_labels).astype('float32')
y_test = np.asarray(test_labels).astype('float32')
```

### Model structure (fully connected feedforward DNN)

Implements DNN  with 2 hidden layers; 

16 neurons in each hidden layers; RelU

Output layer – sigmoid – binary classification

```python
from tensorflow.keras import models, layers

# Initialize the model
model = models.Sequential()

# Input layer and first hidden layer with 16 neurons and ReLU activation
model.add(layers.Dense(16, activation='relu', input_shape=(10000,)))

# Second hidden layer with 16 neurons and ReLU activation
model.add(layers.Dense(16, activation='relu'))

# Output layer with sigmoid activation for binary classification
model.add(layers.Dense(1, activation='sigmoid'))
```

### DNN Model training  for IMDB dataset

```python
# Compile the model with optimizer (RMSprop), 
# loss function (binary_crossentropy ) and error metric (accuracy)

model.compile(optimizer='rmsprop',
              loss='binary_crossentropy',
              metrics=['accuracy'])

# Summary of the model
model.summary()

# Split train data into: 
# validation data (the first 10000 reviews in x_train)
# train data (the remaining reviews in x_train )

x_val = x_train[:10000]
y_val = y_train[:10000]

partial_x_train = x_train[10000:]
partial_y_train = y_train[10000:]

#Fit the model with partial train data and validation data 
#(check model.fit function of keras)

model.fit(partial_x_train, partial_y_train, epochs=10, batch_size=128, validation_data=(x_val, y_val))
```

```python
# Evaluate the model on train data. 
# what are the loss and the accuracy ?

# Do the same for test data and compare

train_loss, train_acc = model.evaluate(x_train, y_train) # less loss more accuracy
test_loss, test_acc = model.evaluate(x_test, y_test)
```

### Overfitting issues - Regularization  L2

regularizers.l1(0.001)

regularizers.l1_l2(l1=0.001, l2=0.001)

### Overfitting issues -  Dropout & EarlyStopping

Aplly a combination of mechanisms to deal with overfitting.

```python
from tensorflow.keras import models, layers
from tensorflow.keras.callbacks import EarlyStopping
import matplotlib.pyplot as plt

hidden = 16  # neurons per layer

# create model with dropout
model_combined = models.Sequential()
model_combined.add(layers.Dense(hidden, activation='relu', input_shape=(10000,)))
model_combined.add(layers.Dropout(0.5))  # Dropout de 50%

model_combined.add(layers.Dense(hidden, activation='relu'))
model_combined.add(layers.Dropout(0.5))  # Dropout de 50%

model_combined.add(layers.Dense(1, activation='sigmoid'))  # Camada de saída

# compile the model
model_combined.compile(optimizer='rmsprop',
                       loss='binary_crossentropy',
                       metrics=['accuracy'])

# Early stopping if validation does not improve for 5 epochs
early = EarlyStopping(monitor='val_loss', min_delta=0, patience=5, verbose=True, mode='auto')

# fit he model
history_combined = model_combined.fit(partial_x_train,
                                      partial_y_train,
                                      epochs=20,
                                      batch_size=512,
                                      validation_data=(x_val, y_val),
                                      callbacks=[early])  # Early Stopping

history_dict = history_combined.history

# loss and accuracy values
loss_values = history_dict['loss']
val_loss_values = history_dict['val_loss']
acc_values = history_dict['accuracy']
val_acc_values = history_dict['val_accuracy']

epochs = range(1, len(loss_values) + 1)

# loss graph
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(epochs, loss_values, 'bo', label='Training Loss')
plt.plot(epochs, val_loss_values, 'r', label='Validation Loss')
plt.title('Training and Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()

# accuracy graph
plt.subplot(1, 2, 2)
plt.plot(epochs, acc_values, 'bo', label='Training Accuracy')
plt.plot(epochs, val_acc_values, 'r', label='Validation Accuracy')
plt.title('Training and Validation Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()

plt.show()

# evaluate train set
train_loss, train_acc = model_combined.evaluate(x_train, y_train)
print(f"Final Training Accuracy: {train_acc:.4f}")

# evaluate test set
test_loss, test_acc = model_combined.evaluate(x_test, y_test)
print(f"Final Testing Accuracy: {test_acc:.4f}")
```
